```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv(r'C:/Users/ASUS/Desktop/Subjects/DataSets/3.12. Example.csv')
raw_data.head(10)
```

```python
x = raw_data.copy()
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
x['clustering'] = kmeans.predict(x)
x

```

```python
plt.scatter(x['Satisfaction'], x['Loyalty'],c = x['clustering'], cmap = 'rainbow')
plt.show()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(raw_data)
x_scaled
```

```python
kmean = KMeans(4)
kmean.fit(x_scaled)
x['clustering'] = kmean.predict(x_scaled)
x
```

```python
plt.scatter(x['Satisfaction'], x['Loyalty'],c = x['clustering'], cmap = 'rainbow')
plt.show()
```
