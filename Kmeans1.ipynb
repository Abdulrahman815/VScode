```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv(r'C:/Users/ASUS/Desktop/Subjects/DataSets/3.12. Example.csv')
df = raw_data.copy()
```

```python
df.count()
```

```python
plt.scatter(df['Satisfaction'], df['Loyalty'])
plt.xlabel("Satisfaction")
plt.ylabel('Loyalty')
```

```python
x = df.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
x['clustering'] = kmeans.predict(x)
```

```python
plt.scatter(x['Satisfaction'], x['Loyalty'], c= x['clustering'], cmap = 'rainbow')
plt.xlabel("Satisfaction")
plt.ylabel('Loyalty')
```

```python
from sklearn.preprocessing import scale
```

```python
x_scaled = scale(x)
x_scaled
```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
x = range(1,10)
plt.plot(x, wcss)
    
```
